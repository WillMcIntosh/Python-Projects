```python
import pandas
df3=pandas.read_csv("/home/will/projects/Udemy_python/files/supermarkets.csv")
df3
```

```python
import geopy
from geopy.geocoders import Nominatim
nom = Nominatim()
```

```python
nom.geocode("3833 Cummins st, Houston, TX 77027")
```

```python
df3["Full Address"]=df3['Address']+", "+df3['City']+", "+df3['State']+", "+df3['Country']
df3
```

```python
df3["Coordinates"]=df3["Address"].apply(nom.geocode)
df3
```

```python
df3["Latitude"]=df3["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df3["Longitude"]=df3["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df3
```
